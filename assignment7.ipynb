```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/co2_emissions_kt_by_country.csv')
print(data.head())
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['value'], marker='o', color='b', linestyle='-', linewidth=2, markersize=5)
plt.title('Global CO2 Emissions Over Time', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('CO2 Emissions (in metric tons)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```
